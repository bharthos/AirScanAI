```python
import pandas as pa
NDA = pa.read_csv("D:\AIT614\Project\data_cleaned.csv")
NDA.head()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

# To predict a variable, we take a subset of the remaining variables
ele = ['NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3']

dpip = make_pipeline(
    SimpleImputer(strategy='median'),
    LinearRegression()
)

# Here, we find the filled and missing values in PM 2.5 Column
pm25Value = NDA.dropna(subset=['PM2_5'])
missed_pm2_5 = NDA[NDA['PM2_5'].isnull()]

# Fit the model on known PM2.5 values
dpip.fit(pm25Value[ele], pm25Value['PM2_5'])

# Predict the missing PM2.5 values
estPM25 = dpip.predict(missed_pm2_5[ele])

# Fill in the missing PM2.5 values in the original dataframe
NDA.loc[NDA['PM2_5'].isnull(), 'PM2_5'] = estPM25


NDA.head(100)
```

```python
# Continue the same process as above to the remaining columns where data is missing
ele = ['PM2_5', 'NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3']

dpip = make_pipeline(
    SimpleImputer(strategy='median'),  
    LinearRegression()
)

pm10Value = NDA.dropna(subset=['PM10'])
missed_pm10 = NDA[NDA['PM10'].isnull()]

dpip.fit(pm10Value[ele], pm10Value['PM10'])

estPM10 = dpip.predict(missed_pm10[ele])

NDA.loc[NDA['PM10'].isnull(), 'PM10'] = estPM10

NDA.head()
```
